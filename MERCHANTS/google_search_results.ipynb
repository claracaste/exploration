```python
def build_link(x, google_search_link = "https://www.google.com/search?q="):
    return google_search_link+ "+".join(x.split(" "))

def add_google_search_links(df):
    """
    add google search link with excel hyperlink 
    """

    # add search by combination of 2 in difficult cases
    col1 = 'original_description_plaid'
    col2 = 'company'
    df['Google search link for ' + col1+ " and "+col2] = df[[col1, col2]
                                                    ].apply(lambda x:  "" if pd.isnull(x[0]) or pd.isnull(x[1]) else 
                                                            '=HYPERLINK("'+build_link(x[0]+' '+x[1])+'";"'+x[0]+' '+x[1]+'")'
                                                              , axis=1 )

    # add search for single columns
    columns = ['description', 'original_description_plaid', 'merchant_name_plaid', 'company']
    for col in columns:
        df['google_'+col] = df[col].apply(lambda x: "" if pd.isnull(x) else 
                                              '=HYPERLINK("'+build_link(x)+'";"'+x+'")'
                                              )
    return df


df_trans2 = add_google_search_links(df_trans)
df_trans2.drop('login_provider_additional_attributes', axis=1).head(1000).to_csv('/Users/claracastellanos/Documents/DATA/MERCHANTS/2024_05_17_links.csv')
import requests
from bs4 import BeautifulSoup
page = requests.get("https://www.google.dz/search?q=see")
soup = BeautifulSoup(page.content)
import re
links = soup.findAll("a")
for link in  soup.find_all("a",href=re.compile("(?<=/url\?q=)(htt.*://.*)")):
    print(re.split(":(?=http)",link["href"].replace("/url?q=","")))





import requests
from BeautifulSoup import BeautifulSoup
page = requests.get("https://www.google.dz/search?q=see")
soup = BeautifulSoup(page.content)
links = soup.findAll("a")

from googlesearch import search   

# to search 
query = "see"

links = []
for j in search(query, tld="co.in", num=10, stop=10, pause=2): 
    links.append(j) 
```
